Метод Симпсона


```python
from scipy.integrate import simpson

def Dubno_simpson(df):
    h = df['X'].iloc[1] - df['X'].iloc[0]
    S = df['Y'].iloc[0] + df['Y'].iloc[len(df)-1]
    for i in range(1, len(df)-1):
        if i % 2 == 0:
            S += 2 * df['Y'].iloc[i]
        else:
            S += 4 * df['Y'].iloc[i]
    return S * h / 3

# Результат вашої функції
result_custom = Dubno_simpson(df)

# Перевірка за допомогою вбудованої функції simpson
result_scipy = simpson(df['Y'], df['X'])

# Виведення результатів
print(f"Результат власної реалізації: {result_custom}")
print(f"Результат за допомогою simpson: {result_scipy}")
```

```python
import pandas as pd
import numpy as np

def Dubno_trapezoid(df1):
    h = df1['X'].iloc[1] - df1['X'].iloc[0]
    S = (df1['Y'].iloc[0] + df1['Y'].iloc[-1]) / 2
    for i in range(1, len(df1)-1):
        S += df1['Y'].iloc[i]
    return S * h  # Враховуємо крок інтегрування для всієї суми

# Результат власної реалізації
result_custom = Dubno_trapezoid(df1)
print(f'Результат власної реалізації: {result_custom}')

# Перевірка за допомогою вбудованої функції numpy.trapz
result_numpy = np.trapz(df1['Y'], df1['X'])
print(f'Результат за допомогою numpy.trapz: {result_numpy}')
```

```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')



# Прочитайте csv файл
file_path = '/content/drive/My Drive/bool.xlsx'

df3 = pd.read_excel(file_path)
df3
```

```python
import matplotlib.pyplot as plt
# Побудова графіка
plt.figure(figsize=(8, 4))
plt.plot(df3['X'], df3['Y'], marker='o', linestyle='-', color='b')

# Налаштування графіка
plt.title('Візуалізація частини кривої, що обмежує площу району')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

# Показати графік
plt.show()
```

```python
# Метод прямокутників

x_values = df3['X']
y_values = df3['Y']

# Обчислення інтегралу методом середніх прямокутників
integral = 0
n = len(x_values) - 1  # кількість інтервалів

for i in range(n):
    # Ширина кожного інтервалу
    h = x_values[i+1] - x_values[i]
    # Середнє значення на інтервалі
    f_avg = (y_values[i] + y_values[i+1]) / 2
    # Площа прямокутника
    integral += h * f_avg

# Виведення тільки інтегралу, обчисленого методом середніх прямокутників
print("Метод середніх прямокутників")
print(f"{integral}")

# Перевірка за допомогою вбудованої функції simpson
from scipy.integrate import simpson
result_scipy = simpson(y=df3['Y'], x=df3['X'])  # Використання ключових аргументів

# Виведення тільки інтегралу, обчисленого за допомогою методу Симпсона
print("Перевірка методом Симпсона")
print(f"{result_scipy}")
```
