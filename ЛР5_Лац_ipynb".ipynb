Розв'язання систем нелінійних рівнянь.

```python
import numpy as np
import matplotlib.pyplot as plt

# Область значень для x та y
x_min, x_max = -2, 1
y_min, y_max = -4, 1
step = 0.01

# Створюємо масиви значень x та y
x, y = np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))

# Рівняння системи
eq1 = np.cos(x+0.5) - y - 2
eq2 = np.sin(y) - 2*x - 1

# Створюємо графік
fig, ax = plt.subplots(figsize=(5, 5))

# Додаємо графік першого рівняння
ax.contour(x, y, eq1, levels=[0], colors='red')

# Додаємо графік другого рівняння
ax.contour(x, y, eq2, levels=[0], colors='blue')

# Налаштування графіка
ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Графік системи рівнянь')

# Додаємо стрілки на кінцях осей
ax.arrow(x_min, 0, x_max - x_min, 0, head_width=0.05, head_length=0.05, fc='black', ec='black')
ax.arrow(0, y_min, 0, y_max - y_min, head_width=0.05, head_length=0.1, fc='black', ec='black')

# Додаємо сітку
plt.grid(True)

# Показуємо графік
plt.show()
```

```python
import numpy as np
from scipy import optimize

# Розв'язок системи рівнянь методом простої ітерації
x0 = -0.95
y0 = -1.1

# задаємо функції
def f1(x):
    return np.cos(x + 0.5) - 2

def f2(y):
    return 0.5 * (np.sin(y) - 1)

# Перевірка методу на збіжність
def iter(x, y, e):
    xn = x
    yn = y
    xn1 = f2(yn)
    yn1 = f1(xn)
    n = 1
    while (abs(xn1 - xn) >= e) or (abs(yn1 - yn) >= e):
        xn = xn1
        yn = yn1
        xn1 = f2(yn)
        yn1 = f1(xn)
        n += 1
    print('Simple iteration:')
    print('x =', xn, '\ny =', yn, '\nThe number of iterations =', n)

iter(x0, y0, 0.0001)

# Задаємо функцію для перевірки
def f3(x):
    return np.cos(x[0] + 0.5) - x[1] - 2, np.sin(x[1]) - 2 * x[0] - 1

# Перевірка розв*язку #Відповіді
s = optimize.root(f3, [x0, y0], method='hybr')

print('Check:', s.x)
```
